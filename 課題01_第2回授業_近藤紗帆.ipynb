```python
while True:
    try:
        h = float(input('身長(cm)：')) / 100
        break
    except:
        print('数字を入力してください．')
        continue
while True:
    try:
        w = float(input('体重(kg)：'))
        break
    except:
        print('数字を入力してください．')
        continue
bmi = w / h**2
print('BMIは，', bmi)
```
