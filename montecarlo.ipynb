```python
import random
import matplotlib.pyplot as plt

#近似値の値を返す関数approx
def approx(sample):
    Nin = 0
    Nout = 0
    for i in range(sample):
        x = random.random() - 0.5
        y = random.random() - 0.5
        if (x**2 + y**2 <= 0.25):
            Nin += 1
        else:
            Nout += 1
    ap = 4*Nin / (Nout + Nin)
    return ap

#メイン
for i in range(1, 7):
    sample = 10 ** i
    ans = approx(sample)
    print('サンプル', str(sample).ljust(14 - i), 'の時のπの近似値:', ans , sep = '')
```

```python
#できなかったやつ・・わ゛か゛ら゛ん゛

import random
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
# mpl.rcParams['axes.autolimit_mode'] = 'round_numbers'
mpl.rcParams['axes.xmargin'] = 0
mpl.rcParams['axes.ymargin'] = 0
xlist = []
ylist = []

#ランダムに設定したx,yを配列に保存する
def xy(sample):
    for i in range(sample):
        x = random.random() - 0.5
        y = random.random() - 0.5
        xlist.append(x)
        ylist.append(y)
    return xlist, ylist

#NinとNoutの値を返す関数check
def check(sample, x, y):
    Nin = 0
    Nout = 0
    for i in range(sample):
        if (x[i]**2 + y[i]**2 <= 0.25):
            Nin += 1
        else:
            Nout += 1
    return Nin, Nout

#近似値の値を返す関数approx
def approx(Nin, Nout):
    ap = 4*Nin / (Nout + Nin)
    return ap

#メイン
for i in range(1, 7):
    sample = 10 ** i
    xlist, ylist = xy(sample)
    Nin, Nout = check(sample, xlist, ylist)
    ans = approx(Nin, Nout)
    print('サンプル', str(sample).ljust(14 - i), 'の時のπの近似値:', ans , sep = '')
    

#プロットの表示
fig = plt.figure(figsize=(8,5), tight_layout = True)
ax1 = fig.add_subplot(231, title="sample=10",
                                    xticks = [-0.5, -0.25, 0, 0.25, 0.5],
                                    yticks = [-0.5,-0.25,0,0.25,0.5])
ax2 = fig.add_subplot(232,title="sample=100",
                                    xticks = [-0.5,-0.25,0,0.25,0.5],
                                    yticks = [-0.5,-0.25,0,0.25,0.5])
ax3 = fig.add_subplot(233, title="sample=1000", 
                                    xticks = [-0.5,-0.25,0,0.25,0.5],
                                    yticks = [-0.5,-0.25,0,0.25,0.5])
ax4 = fig.add_subplot(234, title="sample=10000",
                                    xticks = [-0.5,-0.25,0,0.25,0.5],
                                    yticks = [-0.5,-0.25,0,0.25,0.5])
ax5 = fig.add_subplot(235, title="sample=100000",
                                    xticks = [-0.5,-0.25,0,0.25,0.5],
                                    yticks = [-0.5,-0.25,0,0.25,0.5])
ax6 = fig.add_subplot(236,title="sample=1000000",
                                    xticks = [-0.5,-0.25,0,0.25,0.5],
                                    yticks = [-0.5,-0.25,0,0.25,0.5])

xlist, ylist = xy(10)
#ax1.scatter(xlist, ylist, marker='o')
```
